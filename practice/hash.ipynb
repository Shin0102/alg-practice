# 해쉬 연습

```python
class Node:
    def __init__(self, key, data, next = None):
        self.key = key
        self.data = data
        self.next = next
        
class LinkedList:
    def __init__(self, head: Node = None):
        self.head = head
    
    def add(self, key, data):
        if self.head is None:
            self.head = Node(key, data)
        else:
            node = self.head
            while node.next:
                if node.key == key:
                    break
                node = node.next
            if node.key == key:
                node.data = data
            else:
                print(f"hash collison: {key}, {data}")
                node.next = Node(key, data)

    def delete(self, key):
        node = self.head
        # head
        if node.key == key:
            temp = node
            self.head = node.next
            del temp
            return
        else:
            while node.next:
                if node.next.key == key:
                    temp = node.next
                    node.next = temp.next
                    del temp
                    return
                else:
                    node = node.next
        print(f"{key}와 일치하는 데이터가 없습니다.")
    
    def desc(self):
        node = self.head
        print("------------------------------")
        while node:
            print(f"{node.key}: {node.data}")
            node = node.next
        return ("------------------------------")


import hashlib


class Hash:
    def __init__(self):
        self.hash_size = 8
        self.hash_table = [0 for _ in range(self.hash_size)] 

    def add_data(self, key, data):
        hash_address = self.hash_func(key)
        if self.hash_table[hash_address] == 0:
            self.hash_table[hash_address] = LinkedList(Node(key, data))
        else:
            node = self.hash_table[hash_address] 
            node.add(key, data)

    def delete_data(self, key):
        hash_address = self.hash_func(key)
        if self.hash_table[hash_address] == 0:
            print(f"not found key {key}")
        else: 
            node = self.hash_table[hash_address]
            node.delete(key)

    def hash_func(self, key):
        hash_key = self._hash_key(key)
        return hash_key % self.hash_size

    def _hash_key(self, key):
        """sha256 to interger"""
        hash_key = hashlib.sha256(key.encode())
        return int(hash_key.hexdigest(), 16)
    
        
if __name__ == "__main__":
    hash = Hash()
    print(hash.hash_func("alli"))
    print(hash.hash_func("walter"))
    hash.add_data("alli", "01012341234")
    hash.add_data("alli", "01043214321")
    hash.add_data("walter", "01012345678")
    print(hash.hash_table[3].desc())
    hash.delete_data("alli")
    hash.delete_data("alli2")
    print(hash.hash_table[3].desc())
    
```
