# 트리연습

```python
class Node:
    def __init__(self, data, left = None, right = None):
        self.data = data
        self.left = left
        self.right = right


class TreeManager:
    def __init__(self, head = None):
        self.head = head
        self.parent = head
        self.parent_branch = ""

    def insert(self, data):
        self.current_node = self.head
        if self.current_node is None:
            self.head = Node(data)
            return

        while True:
            if data < self.current_node.data:
                if self.current_node.left is None:
                    self.current_node.left = Node(data)
                    break
                else:
                    self.current_node = self.current_node.left
            else:
                if self.current_node.right is None:
                    self.current_node.right = Node(data)
                    break
                else:
                    self.current_node = self.current_node.right

    def delete(self, data):
        #
        # node search
        is_search = self.search(data)
        if not is_search:
            return False

        #
        # node delete
        #
        # case1 - leaf node
        if self.current_node.left is None and self.current_node.right is None:
            if self.parent_branch == "left":
                self.parent.left = None
            else:
                self.parent.right = None
            print(f"delete case1 {self.current_node.data}")
            del self.current_node
            
        # case2.1 - have one left branch node 
        elif self.current_node.left is not None and self.current_node.right is None:
            if self.parent_branch == "left":
                self.parent.left = self.current_node.left
            else:
                self.parent.right = self.current_node.left
            print(f"delete case2.1 {self.current_node.data}")
            del self.current_node
            
        # case2.2 - have one right branch node 
        elif self.current_node.left is None and self.current_node.right is not None:
            if self.parent_branch == "left":
                self.parent.left = self.current_node.right
            else:
                self.parent.right = self.current_node.right
            print(f"delete case2.2 {self.current_node.data}")
            del self.current_node

        # case3 - have two branch node
        else:
            # case 3.1 - parent branch is left
            if self.parent_branch == "left":
                self.change_node = self.current_node.right
                self.change_node_parent = self.current_node.right
                while self.change_node.left:
                    self.change_node_parent = self.change_node
                    self.change_node = self.change_node.left
                if self.change_node.right:
                    self.change_node_parent.left = self.change_node.right
                else:
                    self.change_node_parent.left = None
                if self.parent != self.current_node: # not root
                    self.parent.left = self.change_node
                self.change_node.left = self.current_node.left
                self.change_node.right = self.current_node.right
                print(f"delete case3.1 {self.current_node.data}")
                del self.current_node
                    
            
            # case 3.2 - parent branch is right
            else:
                self.change_node = self.current_node.right
                self.change_node_parent = self.current_node.right
                while self.change_node.left:
                    self.change_node_parent = self.change_node
                    self.change_node = self.change_node.left
                if self.change_node.right:
                    self.change_node_parent.left = self.change_node.right
                else:
                    self.change_node_parent.left = None
                if self.parent != self.current_node: # not root
                    self.parent.right = self.change_node
                self.change_node.left = self.current_node.left
                self.change_node.right = self.current_node.right
                print(f"delete case3.2 {self.current_node.data}")
                del self.current_node
        return True

    def search(self, data):
        self.current_node = self.head
        self.parent = self.head
        while self.current_node:
            if self.current_node.data == data:
                # print(f"search success {self.current_node.data}")
                return True
            elif data < self.current_node.data:
                self.parent = self.current_node
                self.parent_branch = "left"
                self.current_node = self.current_node.left
            else:
                self.parent = self.current_node
                self.parent_branch = "right"
                self.current_node = self.current_node.right
        return False
        

if __name__ == "__main__":
    import random

    # test set
    bst_nums = set()
    while len(bst_nums) != 15:
        bst_nums.add(random.randint(0, 999))

    print(bst_nums)

    # insert    
    tree = TreeManager()
    for num in bst_nums:
        tree.insert(num)


    print("insert finish")

    # search
    for num in bst_nums:
        if tree.search(num) == False:
            print ('search failed', num)


    print("search finish")

    # delete
    delete_nums = set()
    bst_nums = list(bst_nums)
    while len(delete_nums) != 5:
        delete_nums.add(bst_nums[random.randint(0, 9)])
    print(delete_nums)

    # 선택한 10개의 숫자를 삭제 (삭제 기능 확인)
    for del_num in delete_nums:
        if tree.delete(del_num) == False:
            print('delete failed', del_num)

    print("delete finish")
        
    # tree = TreeManager()
    # tree.insert(50)
    # tree.insert(0)
    # tree.insert(100)
    # tree.insert(75)
    # tree.insert(125)
    # tree.delete(100)

    # print(tree.search(100))
    
            
```
